# **Creating Histogram**

Ploting a histogram of population figures for countries in the year 2020 to get a better sense of the distribution. The histogram will show exactly how many of those countries' population values fall into different population ranges, giving an idea about the spread and concentration. This code will return an obvious and correct visualization by setting the number of bins at 30, removing missing values.

```python
# Select the data for the year 2020
year = '2020'
population_data = data[year].dropna()  # dropna() to remove any missing values

# Create the histogram
plt.figure(figsize=(10, 6))
plt.hist(population_data, bins=30, edgecolor='black')
plt.title('Histogram of Population in 2020')
plt.xlabel('Population')
plt.ylabel('Number of Countries')
plt.show()
```

# **Creating Bar Graph**

A program visualizing the top 10 most populous countries in the year 2020. It first filters and sorts the population data, then makes a bar graph to outline these countries and their respective populations.

```python
# Assume the latest year is 2020 (adjust as needed based on your data)
latest_year = '2020'

# Select the data for the latest year and sort by population
top_10_countries = data[['Country Name', latest_year]].dropna().sort_values(by=latest_year, ascending=False).head(10)

# Create the bar graph
plt.figure(figsize=(12, 8))
plt.bar(top_10_countries['Country Name'], top_10_countries[latest_year], color='skyblue')
plt.title(f'Top 10 Most Populous Countries in {latest_year}')
plt.xlabel('Country')
plt.ylabel('Population')
plt.xticks(rotation=45)
plt.show()
```
